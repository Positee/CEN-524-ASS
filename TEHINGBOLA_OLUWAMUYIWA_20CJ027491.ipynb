```python
import matplotlib.pyplot as plt
import numpy as np
#Generate a sequence of numbers from -10 to 10 with 500 steps 
y = np.linspace(-20,30,600)
#Create a second array z as the cosine of y
z = np.cos(y)
#Use the plot function to plot array y against array z
plt.plot(y,z,marker='x')
```

```python
import pandas as pd
from IPython.display import display

# Create a simple dataset of people using a dictionary
data = {
    'Name': ["Tee", "Posi", "Seyi", "Daniel", "Joy", "Gift", "Tireni" , "Chiwikuem", "Omolara"],
    'State': ["Ogun", "Imo", "Delta", "Kwara", "Lagos", "Abuja", "Abia", "Zaria", "Kaduna"],
    'Age': [18, 20, 26, 28, 16, 17, 90, 15, 47],
    'Height': ["4'1","7'1","4'2","4'3","4'5","4'6","4'7","6'0","5'6"],
    'Gender': ["M","M","M","M","F","F","M","M","F"]
        }

# Convert the dictionary to a pandas DataFrame
data_pandas = pd.DataFrame(data)

# Display the full DataFrame
display(data_pandas)

# Select all rows where the age is less than or equal to 25
display(data_pandas[data_pandas.Age <= 25])
```

```python
## TensorFlow Dataset (TFDS) can be installed with:
!pip install -q tfds-nightly tensorflow
## Import the required libraries 
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
import tensorflow_datasets as tfds

## To get the list of available dataset, use tfds.list_builders()
tfds.list_builders()
##Load the dataset with tfds.load and convert the pandas.DataFrame with tfds.as_dataframe to Visualise the dataset
ds, info = tfds.load('mnist', split='train', shuffle_files=True, with_info=True)
tfds.as_dataframe(ds.take(4), info)
##tfds.show_examples returns a matplotlib.figure to show selected images
##(Note that only image datasets are supported now) 
fig = tfds.show_examples(ds, info)
```
